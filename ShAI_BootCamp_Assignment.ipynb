```python
#import paskages

import pandas as pd
import numpy as np

# for Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

**Basic Data Exploration**

```python
#print name columns
df.columns
```

```python
#get infomation about data
df.info()
```

```python
#print number rows and columns
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# detect  data types columns
print("Data types for each column",df.dtypes)
```

```python
#check missing value
null_data = df.isnull().sum()
null_data
```

**Descriptive Statistics:**

```python
#print a description about all data
df.describe()
```

```python
#describe totalPay salary
desc_totalpay= df['TotalPay'].describe()
print("describtion TotalPay is ",desc_totalpay)
```

```python
#detect range amount salary
range_salary = df['TotalPay'].max() - df['TotalPay'].min()
print("range salary is ",range_salary)
```

```python
#standard deviation for salary
std_dev_salary = df['TotalPay'].std()
print("standard deviation for salary  :",std_dev_salary)
```

**Data Cleaning:**

```python
#feature is contain more 1 null data
null_data[null_data>1]
```

```python
#will check feature Base pay
df['BasePay']
```

```python
#will replace missing data by mean
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
```

```python
df['BasePay'].isnull().sum()
```

```python
# wil will process OvertimePay and OtherPay with replace null data by mean
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
```

```python
#ok, will check feature Benefits
df['Benefits']
```

```python
#it seem many null data , i will show present missin data fot all data
df['Benefits'].isnull().sum()/len(df['Benefits'])*100
```

About a quarter of the data is missing. I will remove it so that it does not affect the rest of the data

```python
#drop it feature Benefits
df.drop(['Benefits'], axis=1, inplace=True)
```

```python
#will check feature Notes missing data
df['Notes']
```

```python
df['Notes'].isnull().sum()/len(df['Notes'])*100
```

```python
# all data in features missing , notusifull i will drop it
df.drop(['Notes'], axis=1 , inplace=True)
```

```python
#will check feature Status missing data
df['Status']
```

```python
df['Status'].isnull().sum()/len(df['Status'])*100
```

```python
# all data in features missing , notusifull i will drop it
df.drop(['Status'], axis=1 , inplace=True)
```

**Basic Data Visualization**

```python
# Create a histogram to visualize the distribution of salaries
plt.hist(df['TotalPay'], bins=20, edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Use pie charts to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(10)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Top 10 Job Titles')
plt.show()
```

```python
#  bar chart to visualize the distribution of employees  different departments

department_counts = df['JobTitle'].value_counts().head(10)
department_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.show()
```

```python
#pair plot to explore relationships between multiple  variables
sns.pairplot(df[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay']])
plt.show()
```

```python
# heatmap to visualize correlations
correlation_matrix = df[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Grouped Analysis:**

```python
# Group the data by 'Year' and calculate the average salary for each year
average_salary_by_year = df.groupby('Year')['TotalPay'].mean()
print("Average Salary by Year:")
print(average_salary_by_year)
```

```python
# Group the data by 'JobTitle' and calculate summary statistics for each group
grouped_data = df.groupby('JobTitle')['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])
grouped_data = grouped_data.sort_values(by='mean', ascending=False).head(10)

print("Statistics for Top 10 Job Titles:")
grouped_data
```

**Simple Correlation Analysis:**

```python
 # correlation between 'TotalPay' and 'BasePay'
correlation = df['TotalPay'].corr(df['BasePay'])
print(f"Correlation between TotalPay and BasePay: {correlation}")

# Plot a scatter plot to visualize the relationship
plt.scatter(df['BasePay'], df['TotalPay'])
plt.title('Scatter Plot: TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
